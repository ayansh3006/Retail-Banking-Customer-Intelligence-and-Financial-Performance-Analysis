```python

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option("display.max_columns", None)

# Better chart style
plt.style.use("ggplot")

print("Libraries Imported Successfully")
```

```python

clients = pd.read_excel("Banking.xlsx", sheet_name="Clients - Banking")
gender = pd.read_excel("Banking.xlsx", sheet_name="Gender")
investment_advisor = pd.read_excel("Banking.xlsx", sheet_name="Investment Advisor")
banking_relationship = pd.read_excel("Banking.xlsx", sheet_name="Banking Relationship")
```

```python
df = clients.merge(gender, on="GenderId", how="left")

df = df.merge(
    banking_relationship,
    on="BRId",
    how="left"
)

df = df.merge(
    investment_advisor,
    on="IAId",
    how="left"
)
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
```

## **Exploratory Data Analysis**

1: What is the age distribution of the bank's customers?

```python
plt.figure(figsize=(8,5))

sns.histplot(df["age"], bins=15, kde=True)

plt.title("Distribution of Customer Age")
plt.xlabel("Age")
plt.ylabel("Number of Customers")

plt.show()
```

Most customers belong to the 75 to 80 years age group, indicating that the bank primarily serves old-age peoples.

2: Gender Distribution

```python
plt.figure(figsize=(6,6))

df["gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)

plt.ylabel("")
plt.title("Gender Distribution")

plt.show()
```

The customer base is fairly balanced between male and female customers.

3: Top Occupations

```python
# Top 10 Occupations by Number of Customers

top_10_occupations = (
    df["occupation"]
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_10_occupations.values,
    y=top_10_occupations.index
)

plt.title("Top 10 Occupations by Number of Customers", fontsize=14)
plt.xlabel("Number of Customers")
plt.ylabel("Occupation")

plt.tight_layout()

plt.show()
```

The bank has the highest number of customers from Associate Proffesor & Structural Analysis Engineer occupation.

4: Estimated Income Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df["estimated_income"], bins=20)

plt.title("Estimated Income Distribution")

plt.show()
```

Customer income is positively skewed, with a small proportion of high-income customers.

5: Top Nationalities by Deposits

```python
deposit_country = (
    df.groupby("nationality")["bank_deposits"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))

deposit_country.plot(kind="bar")

plt.title("Total Bank Deposits by Nationality")

plt.ylabel("Total Deposits")

plt.show()
```

Customers from the top nationality contribute the highest total deposits.

6: Banking Relationship Analysis

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    y="banking_relationship"
)

plt.title("Banking Relationship Distribution")

plt.show()
```

Most customers fall under the Private banking relationship.

7: Loyalty Classification

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="loyalty_classification"
)

plt.title("Loyalty Classification")

plt.show()
```

The majority of customers belong to the Jade loyalty segment.

8: Deposits vs Loans

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="bank_deposits",
    y="bank_loans"
)

plt.title("Deposits vs Loans")

plt.show()
```

Customers with larger deposits generally tend to maintain higher loan balances.

9: Correlation Heatmap

```python
plt.figure(figsize=(12,10))

corr = df.select_dtypes(include="number").corr()

sns.heatmap(
    corr,
    cmap="coolwarm",
    annot=False
)

plt.title("Correlation Matrix")

plt.show()
```

Identify the strongest positive and negative relationships between financial variables.

10: Top 10 Customers by Deposits

```python
top10 = (
    df.nlargest(10, "bank_deposits")
)

plt.figure(figsize=(10,6))

sns.barplot(
    data=top10,
    x="bank_deposits",
    y="name"
)

plt.title("Top 10 Customers by Deposits")

plt.show()
```

These customers represent high-value clients who should receive premium banking services.
